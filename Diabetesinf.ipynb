```python
import pickle
import pandas as pd
```

```python
with open('best_diabetes_model.pkl', 'rb') as f:
    best_model = pickle.load(f)
```

```python
data_inf = {
    'High_BP': [1],
    'High_Chol': [0],
    'Chol_Check': [1],
    'BMI': [27],
    'Smoker': [0],
    'Stroke': [0],
    'Heart_Disease': [0],
    'Physical_Activity': [1],
    'Fruits': [1],
    'Veggies': [1],
    'Alcoholic': [0],
    'Healthcare': [1],
    'No_Doctor_Because_Cost': [0],
    'General_Health': [3],
    'Mental_Health': [2],
    'Physical_Health': [1],
    'Difficulty_Walking': [0],
    'Sex': [1],
    'Age': [9],
    'Education': [4],
    'Income': [6]
}

df_inf = pd.DataFrame(data_inf)
```

```python
prediction = best_model.predict(df_inf)
prediction_proba = best_model.predict_proba(df_inf)
```

```python
if prediction[0] == 1:
    print("Model memprediksi individu BERISIKO diabetes")
    print(f"Probabilitas diabetes: {prediction_proba[0][1]:.2f}")
else:
    print("Model memprediksi individu TIDAK berisiko diabetes")
    print(f"Probabilitas diabetes: {prediction_proba[0][1]:.2f}")
```
